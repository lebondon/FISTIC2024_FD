```python
import pandas as pd
import seaborn as sns
from utils import *
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.linear_model import LinearRegression

PATH="Advertising.csv"

df=pd.read_csv(PATH)

df=df.drop("Unnamed: 0",axis=1)
df
```

```python
df['TV']=df['TV']*1000
df['Radio']=df['Radio']*1000
df['Newspaper']=df['Newspaper']*1000
df['Sales']=df['Sales']*1000
df
```

```python
df.dtypes,df.describe(),df.isnull().sum()
```

```python
TARGET="Sales"
X=df.drop(TARGET,axis=1)
y=df[TARGET]
```

```python
for regressor in X:
    regressor_with_constant=sm.add_constant(X[regressor])
    model=sm.OLS(y,regressor_with_constant).fit()
    display(extract_regression_summary(model))
```
